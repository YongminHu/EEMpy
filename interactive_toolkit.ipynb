```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px


# class EEMInteractivePlot():
    
#     def __init__(self, )

fi_range = (0, 6000)

# Create the Dash app
app = dash.Dash(__name__)

# Define the layout
app.layout = html.Div([
    dcc.Graph(id="heatmap"),
#     dcc.RangeSlider(
#         id="Intensity",
#         min=fi_range[0],
#         max=fi_range[1],
#         step=1,
#         value=[0,4000],  # Initial value for color range
#         marks={int(i): {'label': str(i), 'style':{'fontSize': '15px'}} for i in np.linspace(fi_range[0],fi_range[1],5)},  # Slider labels
#         tooltip={
#                 "placement": "bottom",
#                 "always_visible": True,
#                 "style": {"color": "LightSteelBlue", "fontSize": "15px"},
#     }
#     ),
    dcc.Input(id="intensity_min", type='number', value=0, placeholder="", style={'marginRight':'10px'}),
    dcc.Input(id="intensity_max", type='number', value=6000, placeholder="", style={'marginRight':'10px'}),
])

# Define callback to update heatmap color scale
@app.callback(
    dash.dependencies.Output("heatmap", "figure"),
#     [dash.dependencies.Input("Intensity", "value")]
    dash.dependencies.Input("intensity_min", "value"),
    dash.dependencies.Input("intensity_max", "value")
)
def update_heatmap_color_range(intensity_min, intensity_max):
    # Update the color scale based on the slider value
    
    if intensity_min == None:
        intensity_min = 0
    if intensity_max == None:
        intensity_max = 0

    if intensity_max <= intensity_min:
        intensity_max = intensity_min + 1
    fig = px.imshow(eem_stack[0],
                    labels=dict(x="Emission wavelength (nm)", y="Excitation wavelength (nm)", color="Intensity (a.u.)"),
                    x=em_range,
                    y=np.flip(ex_range),
                    color_continuous_scale='jet',
                    range_color = [intensity_min, intensity_max],
                    origin='lower'
                   )

    return fig

if __name__ == "__main__":
    app.run_server(debug=True)
```
